```python
numeric_columns = data.select_dtypes(include=["int64", "float64"])
num_columns = numeric_columns.shape[1]
cols = 3  
rows = math.ceil(num_columns / cols)  
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(5*cols, 5*rows))
axes = axes.flatten()
for i, col in enumerate(numeric_columns.columns):
    axes[i].boxplot(numeric_columns[col])
    axes[i].set_title(col)
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

```python

def remove_outliers_iqr(data, columns=None):

    if columns is None:
        columns = data.select_dtypes(include=["int64", "float64"]).columns
    
    df = data.copy()
    
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    return df
cleaned_data = remove_outliers_iqr(data)

numeric_columns = cleaned_data.select_dtypes(include=["int64", "float64"])

num_columns = numeric_columns.shape[1]
cols = 3
rows = math.ceil(num_columns / cols)

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(5*cols, 5*rows))
axes = axes.flatten()

for i, col in enumerate(numeric_columns.columns):
    axes[i].boxplot(numeric_columns[col].dropna())  
    axes[i].set_title(col)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
